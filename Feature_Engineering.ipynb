```python
import pandas as pd
```

```python
df = pd.read_csv(r'cleaned_df.csv')
df
```

```python
df.drop(columns=['TransactionDate', 'TransactionTime'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['CustGender'] = encoder.fit_transform(df['CustGender'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df.iloc[:, [2, 5]] = scaler.fit_transform(df.iloc[:, [2, 5]])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(
    transformers=[
        ('encode', CountVectorizer(), 'CustLocation')
    ],
    remainder='passthrough'
)
```

```python
transformed_data = ct.fit_transform(df)
```
